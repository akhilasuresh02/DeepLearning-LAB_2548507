```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, Dense
```

```python
# Load text file
with open("/content/dataset.txt", "r", encoding="utf-8") as file:
    text = file.read()

print(text[:500])
```

## Create Vocabulary and Vectorize Text

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

word_index = tokenizer.word_index
total_words = len(word_index) + 1

print("Total vocabulary size:", total_words)
```

```python
token_list = tokenizer.texts_to_sequences([text])[0]
print(token_list[:20])
```

Create Training Sequences

We create n-gram sequences.

```python
input_sequences = []

for i in range(1, len(token_list)):
    n_gram = token_list[:i+1]
    input_sequences.append(n_gram)

print(input_sequences[:5])
```

Padding Sequences

Neural networks require fixed-length inputs.

```python
max_sequence_len = max([len(seq) for seq in input_sequences])

input_sequences = np.array(
    pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')
)
```

## Split Predictors and Labels

```python
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

y = tf.keras.utils.to_categorical(y, num_classes=total_words)
```

## Build RNN Model

```python
rnn_model = Sequential()

rnn_model.add(Embedding(total_words, 100, input_length=max_sequence_len-1))
rnn_model.add(SimpleRNN(150))
rnn_model.add(Dense(total_words, activation='softmax'))

rnn_model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

rnn_model.summary()
```

## Build LSTM Model

```python
lstm_model = Sequential()

lstm_model.add(Embedding(total_words, 100, input_length=max_sequence_len-1))
lstm_model.add(LSTM(150))
lstm_model.add(Dense(total_words, activation='softmax'))

lstm_model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

lstm_model.summary()
```

```python
history_rnn = rnn_model.fit(
    X, y,
    epochs=15,
    verbose=1
)
```

```python
history_lstm = lstm_model.fit(
    X, y,
    epochs=15,
    verbose=1
)
```

### Text Generation Function


```python
def generate_text(model, seed_text, next_words):

    for _ in range(next_words):

        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences(
            [token_list],
            maxlen=max_sequence_len-1,
            padding='pre'
        )

        preds = model.predict(token_list)[0]
        predicted = np.random.choice(len(preds), p=preds)

        output_word = ""

        for word, index in tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break

        seed_text += " " + output_word

    return seed_text
```

### Generate New Text

```python
print("RNN Generated Text:")
print(generate_text(rnn_model, "artificial intelligence is transforming", 30))
```

```python
print("LSTM Generated Text:")
print(generate_text(lstm_model, "artificial intelligence is transforming", 30))
```

```python
print("Final RNN Accuracy:", history_rnn.history['accuracy'][-1])
print("Final LSTM Accuracy:", history_lstm.history['accuracy'][-1])
```

```python
seed_texts = [
    "artificial intelligence",
    "machine learning systems",
    "technology is transforming",
    "robots are used",
    "deep learning models"
]
```
